```python
def newton_method(f, f_prime, x0, tol=1e-6, max_iter=100):
    """
    Реализация метода Ньютона для нахождения корня уравнения f(x) = 0.

    :param f: функция, корень которой нужно найти
    :param f_prime: производная функции f
    :param x0: начальное приближение
    :param tol: требуемая точность
    :param max_iter: максимальное количество итераций
    :return: приближенное значение корня и количество итераций
    """
    errors_array = []
    x = x0
    for i in range(max_iter):
        fx = f(x)
        fpx = f_prime(x)

        if fpx == 0:
            raise ValueError("Производная равна нулю. Метод Ньютона не применим.")

        x_new = x - fx / fpx
        errors_array.append(2 - x)

        # Проверяем условие достижения требуемой точности
        if abs(x_new - x) < tol:
            return x_new, i + 1, errors_array

        x = x_new

    raise ValueError("Превышено максимальное количество итераций. Метод не сошелся.")

def numerical_derivative(f, h=1e-5):
    """
    Численное дифференцирование функции f.

    :param f: функция, у которой нужно посчитать производную
    :param h: величина, на которую сдвигается значение при численных рассчетах
    :return: функция f'(x), которая принимает один параметр х и возвращает численную производную
    """

    def f_prime(x):
        return (f(x + h) - f(x - h)) / (2 * h)
    return f_prime
```
